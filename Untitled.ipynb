```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from math import e

def f(x,a):
    return a*x**a*e**(-x/a)

X = np.arange(0, 10.25, 0.25)
A = np.arange(1, 2.25, 0.25)
fig, ax = plt.subplots()
data = {}

for a in A:
    y = np.vectorize(lambda x: f(x, a))
    Y = y(X)
    label = f"a = {a}"
    data[label] = Y
    ax.plot(X, Y, label=label)
    
ax.set_xlabel('Значение аргумента X')
ax.set_ylabel('Значении функции У')
ax.set_title('График функции y = $a \cdot x^a \cdot e^{\\frac{-x}{a}}$')
ax.legend()
plt.show()

df = pd.DataFrame(data, index=X)
df
```
